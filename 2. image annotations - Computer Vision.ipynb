### To edit an image making a copy of it is always a good idea!

```python
# Load an image
bflowers = cv2.imread('img/blue-flowers.jpg')
# And make a copy of the original image to avoid manipulating the original image
bflowersimg =bflowers.copy()
imshow(bflowersimg)
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 
cv2.line(bflowersimg,(50,160),(200,80),white,5)
imshow(bflowersimg)
```

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
```

[Hint](# "cv2.addWeighted")  

```python
from PIL import ImageFont,ImageDraw,Image
opacity = 0.8
text = "M and M"
# Load an image
img= cv2.imread('img/m.jpg')
# And make a copy of the original image to avoid manipulating the original image
layer1 =img.copy()
layer2 =img.copy()
cv2.rectangle(layer2,(0,0),(200,75),white,-1)
imgout = cv2.addWeighted(layer2,opacity,layer1,1-opacity,0)
pil = Image.fromarray(imgout)
draw = ImageDraw.Draw(pil)
font =ImageFont.truetype("impact.ttf",30)
draw.text((10,50),text,black,font=font)
processed_img=cv2.cvtColor(np.array(pil),cv2.COLOR_RGB2BGR)
plt.imshow(processed_img)
```
